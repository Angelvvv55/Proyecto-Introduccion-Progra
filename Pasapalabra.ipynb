```python
######Librerias######
import time
from IPython.display import clear_output
import random
import string
from operator import itemgetter

######Funciones######
players = {}

#Registrar jugador.
def reg(name):
  if (name in players) == True:
   print("El jugador {} ya existe.".format(name))
   time.sleep(1)
   clear_output()
  else:
    clear_output()
    opc = int(input("El jugador {} no esta registrado.\n¿Desea registrarlo?\n\n1. Si\n2. No\nRespuesta: ".format(name)))
    if opc == 1:
      players[name] = []
      time.sleep(1)
      clear_output()
      print("Jugador {} registrado.".format(name))
      time.sleep(2)
      clear_output()
    else:
      time.sleep(2)
      clear_output()
      print("\nRegresando al menú principal...\n")
      time.sleep(2)
      clear_output()

#Verificar puntuaciones.
def punt(name):
  clear_output()
  if (name in players) == False:
    clear_output()
    print("El jugador no existe.")
    time.sleep(2)
    clear_output()
  else:
    if players[name]:
     print("Puntuaciones")
     print("Jugador: {}".format(name), players.get(name))
     pues = 0
     playord = dict(sorted(players.items(), key=lambda item:item[1], reverse=True))
     print("Ranking")
     for x in playord:
       playtot = sum(playord[x])
       pues += 1
       if x == name:
        print("\033[1;33m"+("{}. {}: {} puntos.".format(pues,x,playtot)))
       else:
        print("\033[0;30m"+"{}. {}: {} puntos.".format(pues,x,playtot))
     time.sleep(5)
     clear_output()
    else:
      clear_output()
      print("Aun no tiene puntuaciones registradas el jugador ""\033[1;34m" + "{}".format(name))
      time.sleep(2)
      clear_output()

#Eliminar jugador
def borrar(name):
  if (name in players) == False:
   print("No existe el jugador""\033[1;34m" + "{}".format(name))
   time.sleep(1)
  else:
   opc = int(input("¿Esta seguro que desea eliminar al jugador {}?\n\n1. Si\n2. No\nRespuesta: ".format(name, name)))
   if opc == 1:
    del players[name]
    time.sleep(1)
    clear_output()
    print("\nSe eliminó al jugador""\033[1;34m"+ "{}"".".format(name))
    time.sleep(2)
    clear_output()
   else:
    time.sleep(2)
    clear_output()
    print("\nRegresando al menú principal...\n")
    time.sleep(2)
    clear_output()

######Juego######
def game(name):
  clear_output()
  if (name in players) == False:
    clear_output()
    print("El jugador no existe.")
    time.sleep(2)
    clear_output()
  else:
    print("pito")

######Opciones menu######
players = {}
opt = 0
while opt != 5:

  print("\033[1;4;32m"+"Pasapalabras")
  opt = int(input("1.Jugar\n2.Verificar Puntuaciones\n3.Registrar Jugador\n4.Eliminar Jugador\n5.Salir\n"))
  if opt == 1:
    clear_output()
    name = input("Ingrese su nombre de jugador.\n")
    game(name)
  elif opt == 2:
    clear_output()
    name = input("Ingrese su nombre de jugador.\n")
    punt(name)
  elif opt == 3:
    clear_output()
    name = input("Ingrese su nombre de jugador.\n")
    reg(name)
  elif opt == 4:
    clear_output()
    name = input("Ingrese el jugador que desea eliminar.\n")
    borrar(name)
  elif opt == 5:
    clear_output()
  elif opt != (1 or 2 or 3 or 4 or 5):
    print("\033[1;37;41m"+"Introduzca una opción valida.")
    time.sleep(2)
    clear_output()
else:
  print("¡Tenga un buen dia!")
```
